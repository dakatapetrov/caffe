```python
import numpy as np
import pandas as pd
import json

from os.path import expanduser
```

```python
data_path = expanduser("~/data/coco/")
coco_annotation_path = ''.join([data_path, 'Annotations/train2014/'])
annotation_file_pattern = ''.join([coco_annotation_path, '*.json'])
```

```python
def load_csv(path):
    df = pd.read_csv(path)
    present = df[df['present']==1] # filter only present items
    present['object_label'] = present['object_label'].map(lambda x: x.strip())
    return present
```

```python
import glob
anno_files = glob.glob(annotation_file_pattern)

columns = ['image_id', 'category_id']
dtypes = {'image_id':'int', 'category_id':'int'}

dfs = [None] * len(anno_files)
for i in np.arange(len(anno_files)):
    filename = anno_files[i]
    if i % 1000 == 0:
        print("Processed " + str(i) + " files...")
    with open(filename) as json_data:
        data = json.load(json_data)
        image_annotations = [[annotation['image_id'], annotation['category_id']] for annotation in data['annotation']]
        dfs[i] = pd.DataFrame(image_annotations, columns=columns, index=np.arange(len(image_annotations)), dtype='int32')

df =pd.concat(dfs)
```

```python
df[columns] = df[columns].astype(int)
```

```python
# df = df.drop(df.columns[0], axis=1)
df.to_csv(''.join([data_path, 'present_objects.csv']), index=False)
```
